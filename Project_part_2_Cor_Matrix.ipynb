```python
import numpy as np
import pandas as pd
import seaborn as sns
m = pd.read_csv('movies.csv')
g = pd.read_csv('genome-scores.csv')
t = pd.read_csv('tags.csv')
r = pd.read_csv('ratings.csv')
```

### Merge Genome-Scores and Movies data set

```python
# Combining genome-scores.csv and movies.csv into single pandas df.
gm_merge = pd.merge(g, m, 
                    left_on='movieId', 
                    right_on='movieId',
                    how='left')
gm_merge.corr()
```

```python
sns.heatmap(gm_merge.corr(), annot=True)
```

### Merge Ratings and Movies data set

```python
# Combining tags.csv and ratings.csv into single pandas df
rm_merge = pd.merge(r, m, 
                    left_on='movieId', 
                    right_on='movieId',
                    how='left')
```

```python
rm_merge
```

```python
sns.heatmap(rm_merge.corr(), annot=True)
```

```python
rm_merge
```

```python
gm_merge = gm_merge.drop(['title'], axis = 1)
gm_merge = gm_merge.drop(['genres'], axis = 1)
```

```python
gm_merge
```

```python
gm_rm_merge = pd.merge(rm_merge, gm_merge, 
                    left_on='movieId', 
                    right_on='movieId',
                    how='left')
gm_rm_merge.corr()
```
